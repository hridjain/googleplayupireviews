```python
from google_play_scraper import app

import pandas as pd

import numpy as np
```

latest 20000 reviews

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.google.android.apps.nbu.paisa.user',
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=20000, # defaults to 100

)
```

next 20000 latest reviews

```python

resultnext, continuation_token = reviews(
    'com.google.android.apps.nbu.paisa.user',
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=20000, # defaults to 100

)
resultnext, _ = reviews(
    'com.google.android.apps.nbu.paisa.user',
    continuation_token=continuation_token # defaults to None(load from the beginning)
)
```

```python
gpay_new = result + resultnext
```

Making DataFrame, Setting index and exporting as csv

```python
db_gpay_new = pd.DataFrame(gpay_new)
db_gpay_new = db_gpay_new.set_index('reviewId')
db_gpay_new.head()
```

```python

helpful, continuation_token = reviews(
    'com.google.android.apps.nbu.paisa.user',
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.NEWEST
    count=20000, # defaults to 100

)
```

```python

helpfulnext, continuation_token = reviews(
    'com.google.android.apps.nbu.paisa.user',
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.NEWEST
    count=20000, # defaults to 100

)
helpfulnext, _ = reviews(
    'com.google.android.apps.nbu.paisa.user',
    continuation_token=continuation_token # defaults to None(load from the beginning)
)
```

```python
gpay_helpful = helpful + helpfulnext
```

```python
db_gpay_help = pd.DataFrame(gpay_helpful)
db_gpay_help = db_gpay_help.set_index('reviewId')
db_gpay_help.head()
```

```python
from google.colab import files
```

```python
db_gpay_new.to_csv('gpay_new.csv', index=True)
```

```python
db_gpay_help.to_csv('gpay_help.csv', index=True, escapechar='\\')
```
