```python
from google_play_scraper import app

import pandas as pd

import numpy as np
```

latest 20000 reviews

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.phonepe.app',
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=20000, # defaults to 100

)
```

next 20000 latest reviews

```python

resultnext, continuation_token = reviews(
    'com.phonepe.app',
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=20000, # defaults to 100

)
resultnext, _ = reviews(
    'com.phonepe.app',
    continuation_token=continuation_token # defaults to None(load from the beginning)
)
```

```python
phonepe_new = result + resultnext
```

Making DataFrame, Setting index and exporting as csv

```python
db_phonepe_new = pd.DataFrame(phonepe_new)
db_phonepe_new = db_phonepe_new.set_index('reviewId')
db_phonepe_new.head()
```

```python

helpful, continuation_token = reviews(
    'com.phonepe.app',
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.NEWEST
    count=20000, # defaults to 100

)
```

```python

helpfulnext, continuation_token = reviews(
    'com.phonepe.app',
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.NEWEST
    count=20000, # defaults to 100

)
helpfulnext, _ = reviews(
    'com.phonepe.app',
    continuation_token=continuation_token # defaults to None(load from the beginning)
)
```

```python
phonepe_helpful = helpful + helpfulnext
```

```python
db_phonepe_help = pd.DataFrame(phonepe_helpful)
db_phonepe_help = db_phonepe_help.set_index('reviewId')
db_phonepe_help.head()
```

```python
from google.colab import files
```

```python
db_phonepe_new.to_csv('phonepe_new.csv', index=True)
```

```python
db_phonepe_help.to_csv('phonepe_help.csv', index=True, escapechar='\\')
```
